```python
import requests

url = "https://www.kaggle.com/api/v1/datasets/download/redwankarimsony/heart-disease-data"
headers = {"User-Agent": "Mozilla/5.0"}  # optional

response = requests.get(url, headers=headers, stream=True)

with open("heart-disease-data.zip", "wb") as f:
    for chunk in response.iter_content(chunk_size=8192):
        f.write(chunk)
```

```python
import zipfile
import os

# Path to your downloaded file
zip_path = "/workspaces/heart-disease-prediction-system/heart-disease-data.zip"
extract_to = "/workspaces/heart-disease-prediction-system/heart-disease-data"

# Make sure the folder exists
os.makedirs(extract_to, exist_ok=True)

# Extract all files
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_to)

print("✅ Files extracted to:", extract_to)
```
